```python
def logreg(z):
    y = 1/(1 + np.exp(-z))
    return y
```

### softmax
#### - 다중분류에서 사용하는 최적화함수로 다중항목의 결과값을 확률로 치환하여 가장 높은 확률을 가진 항목으로 분류
#### - 마이너스 값을 양수로 치환하기 위해 np.exp를 사용해 결과값을 변경하고 모든 값을 더한 값에 자기 자신값을 나눠서 확률을 구한다

```python
score = [-1.5, 2, 1.8]
```

```python
prob = []
sum = 0

for i in range(len(score)):
    prob.append(np.exp(score[i]))
    sum = sum + np.exp(score[i])

prob = prob / sum
prob.round(3)                       
```

```python
def softmax(x):
    ex = np.exp(x - max(x))
    return ex / ex.sum()
```

```python
x = np.array([1, 1, 2])
y = softmax(x)
```

### 유방암 데이터를 이용하여 여러 알고리즘에 적용해 성능비교 
##### 데이터 : https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data/data 
##### 종속변수 : diagnosis M : 암, B : 정상

```python
data = pd.read_csv('./breast_cancer_data.csv')
data.head(3)
```

```python
data.info()
```

```python
data.shape
```

### 범주형데이터를 수치형으로 변환
#### M->1, B->0
#### map함수는 데이터의 값을 치환시 사용
##### - 데이터.map(함수)


```python
data['diagnosis'] = data['diagnosis'].map({'M':1, 'B':0})
```

```python
df = data[['diagnosis','radius_mean','texture_mean']]
```

```python
features = ['radius_mean','texture_mean']
```

```python
x = df[features]
y = df['diagnosis']
```

```python
np.random.seed(11)
```

```python
train_x, test_x, train_y, test_y = train_test_split(x,y, test_size=0.3)
train_x.shape, test_x.shape, train_y.shape, test_y.shape
```

```python
model = LogisticRegression()
```

```python
model.fit(train_x, train_y)
```

```python
print("score : {0:.2%}".format(model.score(test_x, test_y)))
```

### 입력(특성)데이터를 시각화

```python
mark = ['o','v']
label = ['M','B']

for i in range(2):
    xs = train_x['radius_mean'][train_y == i]
    ys = train_x['texture_mean'][train_y ==i]
    plt.scatter(xs, ys, marker = mark[i], s=25)
plt.legend(label)
plt.xlabel('radius_mean')
plt.ylabel('texture_mean')
plt.show()
```
