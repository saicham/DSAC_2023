```python
fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
rect = plt.Rectangle((0.2,0.75), 0.5, 0.2)
circ = plt.Circle((0.7,0.2), 0.15)

ax.add_patch(rect)
ax.add_patch(circ)
```

```python
c1 = np.random.normal(0,1, size=200)
c2 = np.random.normal(10,2, size=200)
np.column_stack(c1,c2)
```

```python
#   
df1 = pd.DataFrame(np.column_stack((c1,c2)))
```

```python
#    pd.concat 
df2 = pd.Series(np.concatenate([c1,c2]))
```

```python
df2.hist(bins=100,density=True)
df2.plot(kind='kde',style='r--')
# plt.text(0.1,0.5,'test text')
plt.text(df2.mean(), 0.5, 'test text', ha='center', va='center')
```
