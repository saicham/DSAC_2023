### JSON to DataFrame
#### json_normalize() 를 사용하여 JSON 데이터를 DataFrame로 전환 

```python
# 수집한 데이터를 json구조로 변환
jo = json.loads(r.text)
```

```python
df = json_normalize(jo,'list')  #json데이터의 list를  dataframe으로 변환
len(df)
```

```python
df.head(3)
```

```python
df.columns
```

```python
df = df[['s_name', 'lat', 'lot', 'sido_name', 'gugun_name', 'doro_address', 'tel']]
```

```python
df.dtypes
```

```python
df['lat'] = df['lat'].astype(float)
df['lot'] = df['lot'].astype(float)
df.dtypes
```
