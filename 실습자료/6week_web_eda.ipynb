```python
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x, y = np.sin(x + i * 0.5) * (7-i) * flip))
```

```python
sns.set_style('dark')
plt.subplot(2,1,1)
sinplot()
    
# plt.subplot(2,1,2)
# sinplot(-1)
```

#### 범주형 데이터 처리 

```python
sample_num = 10
```

```python
tight = 170 + 3 * np.random.randn(sample_num).round()
tight
```

```python
natn = np.random.randint(0,3,sample_num)
natn
```

```python
tot_df = pd.DataFrame(list(zip(tight,natn)), columns=['tight','natn'])
tot_df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
nat = pd.get_dummies(tot_df['natn'],prefix='natn_')
nat
```

```python
new_tot = pd.concat([tot_df,nat], axis=1)
new_tot
```

```python
new_tot = new_tot.drop('natn', axis=1)
new_tot
```

```python
pd.Categorical(natn)
```

```python
pd.Categorical(new_tot)
```
