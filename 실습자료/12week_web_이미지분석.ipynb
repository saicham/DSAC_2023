```python
img3 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img3, cmap='gray')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
img4 = cv2.flip(img,1)
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img4, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img5 = cv2.resize(img, (400,300))
plt.imshow(cv2.cvtColor(img5, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img6 = img[0:100, 0:2000]
plt.imshow(cv2.cvtColor(img6, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img7 = cv2.blur(img,(100,100))
plt.imshow(cv2.cvtColor(img7, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
kernel =np.array([[0, -1, 0],
                 [-1, 5, -1],
                 [0, -1,0]])
```

```python
img8 = cv2.filter2D(img, -1, kernel)
plt.imshow(cv2.cvtColor(img8, cv2.COLOR_BGR2GRAY), cmap='gray')
plt.show()
```
